```python
# load data.csv file
import pandas as pd

df = pd.read_csv('data.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a bar chart for each product
plt.bar(df['Month'], df['Product_A'], color='b', alpha=0.6, label='Product A')
plt.bar(df['Month'], df['Product_B'], color='r', alpha=0.6, label='Product B')
plt.bar(df['Month'], df['Product_C'], color='g', alpha=0.6, label='Product C')

# Set the chart title and labels
plt.title('Comparación de las ventas mensuales de cada producto')
plt.xlabel('Month')
plt.ylabel('Quantity')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
